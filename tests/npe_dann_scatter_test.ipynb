```python
from haloflow.npe import valid as V
from haloflow.npe import plotting as P
```

```python
train_sim = test_sim = 'Simba100'
obs = 'mags'
```

```python
ranks, alpha, ecp, y_nde = V.validate_npe(train_obs=obs, train_sim=train_sim, test_obs=obs, test_sim=test_sim, version=1, data_dir='../../data/hf2/dann/npe')
ranks_dann, alpha_dann, ecp_dann, y_nde_dann = V.validate_npe(train_obs=obs, train_sim=train_sim, test_obs=obs, test_sim=test_sim, version=1, with_dann=True, data_dir='../../data/hf2/dann/npe')
```

```python
fig = P.plot_rank_statistics([ranks, ranks_dann], labels=['only NPE', 'NPE + DANN'])
```

```python
fig, ax = P.plot_coverage([alpha, alpha_dann], [ecp, ecp_dann], labels=['only NPE', 'NPE + DANN'])
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(6, 8))

ax, _, _ = P.plot_true_pred(
    ax,
    train_obs=obs,
    train_sim=train_sim,
    test_obs=obs,
    test_sim=test_sim,
    device='cpu',
    mass='halo',
    with_dann=False,
    data_dir='../../data/hf2/dann/npe'
)

ax, _, _ = P.plot_true_pred(
    ax,
    train_obs=obs,
    train_sim=train_sim,
    test_obs=obs,
    test_sim=test_sim,
    device='cpu',
    mass='halo',
    with_dann=True,
    data_dir='../../data/hf2/dann/npe', 
    fmt='.C1'
)

ax.legend(['only NPE', 'NPE + DANN'])

plt.show()
```
